### 단순 연결리스트 수업자료 응용예제 2  - 로또 추첨하기

```python
class Node():
    def __init__(self):
        self.data = None
        self.link = None

def makeLinkedLotto(number):
    global memory,head,current,pre

    

    if number not in memory:
        node = Node()
        node.data = number
        memory.append(node)

        if head == None:
            head = node
            return
        if head.data > number:
            node.link = head
            head  = node
            return

        current = head
        while current.link != None:

            pre = current
            current = current.link

            if current.data > number:
                pre.link = node
                node.link = current
                return

        current.link = node

def printLotto(start):
    global memory,head,current,pre

    if head == None:
        return
    start = head
    while start.link != None:
        print(start.data,end = " ")    
        start = start.link
    print()    
      
import random
memory = []
head,pre,current = None,None,None

if __name__ == "__main__":
    

    while len(memory) != 7 : # head 까지 포함 
        num = random.randint(1,46)
        makeLinkedLotto(num)

    printLotto(head)
```

### 원형연결리스트 수업예제 1  - 현재 위치부터 가까운 편의점 관리하기

```python
import math,random

class Node():
    def __init__(self):
        self.data = None
        self.link = None


def makeCV(cvdata):
    global memory,head,pre,current

    node = Node()
    node.data = cvdata
    memory.append(node)

    if head == None:
        head = node
        node.link = head
        return

    head_x,head_y = head.data[1:]
    node_x,node_y = node.data[1:]

    head_distance = math.sqrt((head_x*head_x)+(head_y*head_y))
    node_distance = math.sqrt((node_x*node_x)+(node_y*node_y))

    if head_distance > node_distance:
        node.link = head
        
        last = head
        while last.link != head:
            last = last.link
        last.link = node
        head = node
        return

    current = head
    while current.link != head:
        pre = current
        current = current.link

        current_x,current_y = current.data[1:]
        current_distance = math.sqrt((current_x*current_x)+(current_y*current_y))
        if current_distance > node_distance:
            pre.link = node
            node.link = current
            return

    current.link = node
    node.link = head        

def printCV(start):
    global memory,head,pre,current
    
    if start.data == None:
        return

    current = start
    while current.link != head:
        current_x,current_y = current.data[1:]
        current_distance = math.sqrt((current_x*current_x)+(current_y*current_y))
        print("편의점 ",current.data[0],"의 거리: ", current_distance)
        current = current.link


memory = []
head,current,pre = None,None,None

if __name__ == "__main__":

    name = 'A'
    for i in range(10):
        cv_data = (name,random.randint(1,100),random.randint(1,100))
        makeCV(cv_data)
        name = chr(ord(name)+1)
    printCV(head)     
```

### 스택 수업자료 응용예제 2 - 파일내용을 거꾸로 출력하기

```python
def isStackFull():
    global SIZE,stack,top
    if top >= SIZE-1:
        print("스택이 꽉 찼음")
        return True
    else:
        return False

def isStackEmpty():
    global SIZE,stack,top        
    if top == -1:
        
        return True
    else :
        return False

def push(data):
    global SIZE,stack,top
    if(isStackFull()):
        return
    else:
        top += 1
        stack[top] = data
        

def pop():
    global SIZE,stack,top 
    if(isStackEmpty()):
        return
    else:
        data = stack[top]
        stack[top] = None
        top -= 1
        return data    

def peek():
    global SIZE,top,stack
    if(isStackEmpty()):
        return
    else:
        return stack[top]

ss = '''
    진달래꽃
    나 보기가 역겨워
    가실 때에는
    말없이 고이 보내드리오리다.
'''   

SIZE = 56
top = -1
stack = [None for _ in range(SIZE)] # SIZE만큼 스택 채워넣기

if __name__ == "__main__":
    print("---- 원본 ----")

    for i in ss:
        push(i)
        print(i,end = " ")
    print()
    
    print("---- 거꾸로 처리된 결과 ----")

    while (isStackEmpty() != True):
        print(pop(),end = " ")
```
