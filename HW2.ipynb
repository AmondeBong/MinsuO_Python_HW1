### 단순 연결리스트 수업자료 응용예제 2  - 로또 추첨하기

```python
import random
class Node():
    def __init__(self):
        self.data = None;
        self.link = None;

def printNodes(start):
        current = start
        if current == None :
            return
        print(current.data,end = ' ')
        while current.link != None :
            current = current.link
            print(current.data,end = ' ')
        print()

def makeSimpleLinkedList(nameEmail):
    global memory,head,current,pre
    
    node = Node()
    node.data = nameEmail
    memory.append(node)

    #첫번째 노드일때
    if head == None:
        head = node
        return
    #첫번째 노드보다 작을때    
    if head.data[1] > nameEmail[1]: 
        node.link  = head
        head = node
        return

    #중간 노드에 넣는경우
    current = head
    while current.link != None:
        pre = current
        current = current.link
        if current.data[1]> nameEmail[1]: # 집어 넣으려는데이터보다 현재 데이터가 큰 경우(앞에 집어넣음)
            pre.link = node
            node.link = current
            return


    current.link = node # 뒤에 아무것도 없는경우

memory = []
head,current,pre = None,None,None 
lotto = []

while True:
    pick_num = random.randint(1,46)
    if pick_num not in lotto:
        lotto.append(pick_num)
    if len(lotto) >= 6:
        break;

lotto.sort()
makeSimpleLinkedList(lotto)
printNodes(head)
```

### 스택 수업자료 응용예제 2 - 파일내용을 거꾸로 출력하기

```python
def isStackFull():
    global SIZE,stack,top
    if top >= SIZE-1:
        print("스택이 꽉 찼음")
        return True
    else:
        return False

def isStackEmpty():
    global SIZE,stack,top        
    if top == -1:
        
        return True
    else :
        return False

def push(data):
    global SIZE,stack,top
    if(isStackFull()):
        return
    else:
        top += 1
        stack[top] = data
        

def pop():
    global SIZE,stack,top 
    if(isStackEmpty()):
        return
    else:
        data = stack[top]
        stack[top] = None
        top -= 1
        return data    

def peek():
    global SIZE,top,stack
    if(isStackEmpty()):
        return
    else:
        return stack[top]

ss = '''
    진달래꽃
    나 보기가 역겨워
    가실 때에는
    말없이 고이 보내드리오리다.
'''   

SIZE = 56
top = -1
stack = [None for _ in range(SIZE)] # SIZE만큼 스택 채워넣기

if __name__ == "__main__":
    print("---- 원본 ----")

    for i in ss:
        push(i)
        print(i,end = " ")
    print()
    
    print("---- 거꾸로 처리된 결과 ----")

    while (isStackEmpty() != True):
        print(pop(),end = " ")
```
